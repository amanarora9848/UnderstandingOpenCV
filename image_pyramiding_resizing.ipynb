The below code will be used for implementaing the image pyramiding operation, which scles the image to half its size, when shrinking, and expands it to double its size, when enlarging, but results in degrade of quality. Useful often in object detection.

```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
from google.colab import files
uploaded = files.upload()
```

```
image = cv2.imread('input.jpg')
plt.imshow(image)
```

The shrink and enlarge are the respective scaling operations.

```
shrink = cv2.pyrDown(image)
plt.imshow(shrink)
```

```
enlarge = cv2.pyrUp(image)
plt.imshow(enlarge)
```

The same can be implemented, if needed, in a loop, for multiple shrinking or enlarging, as per requirement.
