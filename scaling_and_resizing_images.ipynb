For the scaling and resizing functions, we use the cv2.resize(image, dsize(output image size), x_scale_factor, y_scale_factor, interpolation) function. Interpolation re of various types, for eg; linear, INTER_CUBIC, INTER_AREA, etc. Check documentation for the same. useful link: http://tanbakuchi.com/posts/comparison-of-openv-interpolation-algorithms/

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
image = cv2.imread('input.jpg')
plt.imshow(image)
```

```
image_scaled = cv2.resize(image, None, fx = 0.75, fy = 1)
# Default interpolation selected is LINEAR.
plt.imshow(image_scaled)
```

```
image_scaled = cv2.resize(image, None, fx = 2, fy = 2, interpolation = cv2.INTER_CUBIC)
plt.imshow(image_scaled)
```

Here, we fix the heght and width of the image by ourselves.

```
image_scaled = cv2.resize(image, (900, 400), interpolation = cv2.INTER_AREA)
plt.imshow(image_scaled)
```
