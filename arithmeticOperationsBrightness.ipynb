```
image = cv2.imread('input.jpg')
plt.imshow(image)
```

Now, we'll be creating a matix of ones, and then multiply it by a scaler of 100, which will return a matrix the same dimensions of our image with all values being hundred (out of the values of rgb (0-255)).

```
M = np.ones(image.shape, dtype = 'uint8')*160
```

```
added = cv2.add(image, M)
plt.imshow(added)
```

```
subtracted = cv2.subtract(image, M)
plt.imshow(subtracted)
```
