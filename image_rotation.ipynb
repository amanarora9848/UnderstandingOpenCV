Performing rotation operations on images in openCV.
It uses  rotation matrix, with angle of rotation (theta). OpenCV allows us to scale and rotate at the same time using the cv2.getRotationMatrix2D(rotation_center_x, rotation_ecnter_y, angle_of_rotation, scale) function.

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
image = cv2.imread('input.jpg')
height, width = image.shape[:2] # Slicing
```

Now, we'll divide the height and width by two to rotate about the centre.

```
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.5)
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
```

Remember that the angle of rotation, i.e. the value 90ยบ that we've given here, is mesured anti-clockwise.

```
plt.imshow(rotated_image)
```
