OpenCV has o specific cropping function. We rather just use numpy slicing operatios, to slice the image array, to get the portions of the images we want. hence the code.

```
image = cv2.imread('input.jpg')
plt.imshow(image)
```

```
height, width = image.shape[:2]
```

So, we get the start row and column first, 25% of total length, similarly the end row and column as 75% of the total length of image.

```
start_row, start_col = int(height*0.25), int(width*0.25)
end_row, end_col = int(height*0.75), int(width*0.75)
```

Now, we use numpy for cropping.

```
cropped_image = image[start_row:end_row, start_col:end_col]
plt.imshow(cropped_image)
```
